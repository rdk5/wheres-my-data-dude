```python
import pandas as pd
import numpy as np
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
from datetime import date, timedelta
```

```python
list_ = []
weekly_sentiment = []
compound_sentiment = []
weekly_tweets = []
for x in range(26):
    file = "Data/Born a Crime Data/Born a Crime" + str(x) + ".csv"
    df = pd.read_csv(file, encoding="latin-1", error_bad_lines=False, warn_bad_lines=False, index_col=None, header=0)
    
    df["Tweet Sentiment"] = ""
    df.set_index(" id", inplace=True)
    #df
    
    for index, row in df.iterrows():
        tweet_text = row[" text"]
        compound_sentiment.append(analyzer.polarity_scores(tweet_text)["compound"])
        df.set_value(index, "Tweet Sentiment", (analyzer.polarity_scores(tweet_text)["compound"]))

    compound_sentiment = []   
    list_.append(df)
frame = pd.concat(list_)
frame.head()
frame[' date'] = pd.to_datetime(frame[' date'])
```

```python
dates = "Dates/bac_dates.csv"
bac_dates = pd.read_csv(dates)
bac_dates
```

```python
weekly_sentiment = []
weekly_tweets = []
for index, row in bac_dates.iterrows():
    date = row["Week Start"]
    date2 = row["Week End"]
    mask = (frame[' date'] > str(date)) & (frame[' date'] <= str(date2))
    weekly_tweets.append(len(frame.loc[mask]))
    weekly_sentiment.append(np.mean(frame.loc[mask]["Tweet Sentiment"]))
```
