```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn
```

```python
modern_romance = "Full Data CSVs/modern_romance_clean_data.csv"
modern_romance_df = pd.read_csv(modern_romance)
modern_romance_df.head()
```

```python
instant_mom = "Full Data CSVs/instant_mom_clean_data.csv"
instant_mom_df = pd.read_csv(instant_mom)
instant_mom_df.head()
```

```python
back_tattoo = "Full Data CSVs/back_tattoo_clean_data.csv"
back_tattoo_df = pd.read_csv(back_tattoo)
back_tattoo_df.head()
```

```python
bossypants = "Full Data CSVs/bossypants_clean_data.csv"
bossypants_df = pd.read_csv(bossypants)
bossypants_df.head()
```

```python
bornacrime = "Full Data CSVs/born_a_crime_clean_data.csv"
bornacrime_df = pd.read_csv(bornacrime)
bornacrime_df.head()
```

```python
plt.figure(figsize=(15, 7))
x_axis = np.arange(len(modern_romance_df["Weekly Tweets"]))

xticks = ["6 Weeks Before Publication", ]

plt.plot(x_axis, modern_romance_df["Weekly Tweets"], "y", linestyle="dashed", label="Modern Romance Tweets")
plt.plot(x_axis, instant_mom_df["Weekly Tweets"], "g", linestyle="dashed", label ="Instant Tweets")
plt.plot(x_axis, back_tattoo_df["Weekly Tweets"], "k", linestyle="dashed", label="The Girl with the Back Tattoo Tweets")
plt.plot(x_axis, bossypants_df["Weekly Tweets"], "b", linestyle="dashed", label="Bossypants Tweets")
plt.plot(x_axis, bornacrime_df["Weekly Tweets"], "r", linestyle="dashed", label="Born a Crime Tweets")

plt.legend(loc="best", fancybox=True)

plt.show()
```

# Tweets vs. Sales

### Modern Romance

```python
fig, ax1 = plt.subplots(figsize=(15,7))
t = np.arange(len(modern_romance_df["Weekly Tweets"]))
s1 = modern_romance_df["Weekly Sales"]
ax1.bar(t, s1)
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('Sales', color='b', fontsize=14)
#ax1.grid(None)
ax1.tick_params('y', colors='b')
ax1.set_xticks(ticks=t-0.6)
ax1.set_xticklabels(labels=("6 Weeks Out", "5 Weeks Out", "4 Weeks Out", "3 Weeks Out", "2 Weeks Out", "1 Week Out", " Publication Week",
              "1 Week After", "2 Weeks After","3 Weeks After","4 Weeks After","5 Weeks After","6 Weeks After","7 Weeks After",
              "8 Weeks After","9 Weeks After","10 Weeks After","11 Weeks After","12 Weeks After","13 Weeks After","14 Weeks After",
               "15 Weeks After","16 Weeks After"), rotation=45, size="medium"
              )
ax2 = ax1.twinx()
ax2.grid(None)
s2 = modern_romance_df["Weekly Tweets"]
ax2.plot(t, s2, color='r')
ax2.set_ylabel('Weekly Tweets', color='r', fontsize=14)
ax2.tick_params('y', colors='r')

fig.tight_layout()
plt.savefig("ModernRomanceTvS.png")
plt.show()
```
