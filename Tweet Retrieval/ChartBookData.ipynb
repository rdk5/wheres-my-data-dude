### Tweets

```python
plt.figure(figsize=(15, 7))
x_axis = np.arange(len(modern_romance_df["Weekly Tweets"]))

labels = ["6 Weeks Out", "5 Weeks Out", "4 Weeks Out", "3 Weeks Out", "2 Weeks Out", "1 Week Out", " Publication Week",
              "1 Week After", "2 Weeks After","3 Weeks After","4 Weeks After","5 Weeks After","6 Weeks After","7 Weeks After",
              "8 Weeks After","9 Weeks After","10 Weeks After","11 Weeks After","12 Weeks After","13 Weeks After","14 Weeks After",
               "15 Weeks After","16 Weeks After"]

plt.plot(x_axis, modern_romance_df["Weekly Tweets"], "y", linestyle="dashed", label="Modern Romance Tweets")
plt.plot(x_axis, instant_mom_df["Weekly Tweets"], "g", linestyle="dashed", label ="Instant Tweets")
plt.plot(x_axis, back_tattoo_df["Weekly Tweets"], "k", linestyle="dashed", label="The Girl with the Back Tattoo Tweets")
plt.plot(x_axis, bossypants_df["Weekly Tweets"], "b", linestyle="dashed", label="Bossypants Tweets")
plt.plot(x_axis, bornacrime_df["Weekly Tweets"], "r", linestyle="dashed", label="Born a Crime Tweets")

plt.ylabel("Tweets", fontsize=14)

plt.xticks(x_axis-0.6, labels, rotation=45, size="medium")

plt.legend(loc="best", fancybox=True)
plt.savefig("Tweets.png")
plt.show()
```

# Reviews vs. Average Sentiment

```python
review_scores = "average_rating_data.csv"
review_scores_df = pd.read_csv(review_scores)
review_scores_df.set_index("Title")
```

```python
plt.figure(figsize=(7, 7))

plt.plot(modern_romance_df["Weekly Sentiment"].mean(), review_scores_df.iloc[3]["GoodReads_Average_Rating"], "o", markersize=10, label="Modern Romance")
plt.plot(instant_mom_df["Weekly Sentiment"].mean(), review_scores_df.iloc[2]["GoodReads_Average_Rating"], "d", markersize=10, label="Instant Mom")
plt.plot(back_tattoo_df["Weekly Sentiment"].mean(), review_scores_df.iloc[9]["GoodReads_Average_Rating"], "s", markersize=10, label="Back Tattoo")
plt.plot(bossypants_df["Weekly Sentiment"].mean(), review_scores_df.iloc[6]["GoodReads_Average_Rating"], "*", markersize=10, label="Bossypants")
plt.plot(bornacrime_df["Weekly Sentiment"].mean(), review_scores_df.iloc[8]["GoodReads_Average_Rating"], "^", markersize=10, label="Born a Crime")
plt.plot( [0.55, -0.2],[3.7,4.5] )

plt.legend(loc="best", fancybox=True)

plt.savefig("RvAS.png")

plt.show()
```
